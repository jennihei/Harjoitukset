```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
from matplotlib.ticker import MultipleLocator
from matplotlib.dates import DateFormatter
myLoc = MultipleLocator(14)
myFmt = DateFormatter('%d/%m')
```

```python
df1 = pd.read_excel('2018_start.xlsx')
```

```python
df2 = pd.read_excel('2018_stop.xlsx')
```

```python
df1.head()
```

```python
df2.head()
```

```python
#aikaleimat indeksiin
df1.index = pd.to_datetime(df1['Päivämäärä'])
df2.index = pd.to_datetime(df2['Päivämäärä'])

#järjestys vanhimmasta uusimpaan
df1.sort_index(inplace = True)
df2.sort_index(inplace = True)
```

```python
df1.head()
```

```python
df1['Aloitus'].plot()
```

```python
df2['Lopetus'].plot()
```

```python
#luon figure- ja axes-oliot
fig, ax = plt.subplots(figsize = (20, 6))

#määritän ax1-olion
ax.set_ylabel('Aloitetut sopimukset 2018', color='C0', fontsize=14)
ax.plot(df1['Aloitus'], color='C0')
ax.set_ylim(0,20)

ax_T = ax.twinx()

ax_T.set_ylabel('Lopetetut sopimukset 2018', color='C1', fontsize=14)
ax_T.plot(df2['Lopetus'], color='C1')
ax_T.set_ylim(0,20)
```
