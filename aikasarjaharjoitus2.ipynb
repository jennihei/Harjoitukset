```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
from matplotlib.ticker import MultipleLocator
from matplotlib.dates import DateFormatter
myLoc = MultipleLocator(14)
myFmt = DateFormatter('%d/%m')

df1 = pd.read_excel('2019alku.xlsx')
df2 = pd.read_excel('2019loppu.xlsx')
```

```python
df2.head()
```

```python
df1.index = pd.to_datetime(df1['Päivä'])
df2.index = pd.to_datetime(df2['Päivä'])
df1.sort_index(inplace = True)
df2.sort_index(inplace = True)
```

```python

fig, ax = plt.subplots(figsize = (20, 6))


ax.set_ylabel('Aloitetut sopimukset 2019', color='C0', fontsize=14)
ax.plot(df1['Määrä'], color='C0')
ax.set_ylim(0,20)

ax_T = ax.twinx()

ax_T.set_ylabel('Lopetetut sopimukset 2019', color='C1', fontsize=14)
ax_T.plot(df2['Määrä'], color='C1')
ax_T.set_ylim(0,20)
```
